```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
```

```python
plt.style.use(['science', 'notebook', 'grid'])
```

Saved the results for each central pressure on .dat files. After loading them we plot the results.

```python
press_arr = np.genfromtxt('press_data.dat', dtype=float, skip_header=1)
freq_arr = np.genfromtxt('freq_data.dat', dtype=float, skip_header=1)
stab_arr = np.genfromtxt('stab_data.dat', dtype=float, skip_header=1)
ang_arr = np.arcsinh((2*np.pi*freq_arr)**2)
eigen_arr = ang_arr * stab_arr

press1_arr = np.genfromtxt('press1_data.dat', dtype=float, skip_header=1)
press2_arr = np.genfromtxt('press2_data.dat', dtype=float, skip_header=1)
freq1_arr = np.genfromtxt('freq1_data.dat', dtype=float, skip_header=1)
stab1_arr = np.genfromtxt('stab1_data.dat', dtype=float, skip_header=1)
freq2_arr = np.genfromtxt('freq2_data.dat', dtype=float, skip_header=1)

ang1_arr = np.arcsinh((2*np.pi*freq1_arr)**2)
ang2_arr = np.arcsinh((2*np.pi*freq2_arr)**2)
eigen1_arr = ang1_arr * stab1_arr
eigen2_arr = ang2_arr
```

```python
plt.plot(press_arr, eigen_arr, zorder=3, label='$\omega_0^2$')
plt.plot(press1_arr, eigen1_arr, color='tab:orange', linestyle='dashed', zorder=2, label='$\omega_1^2$')
plt.plot(press2_arr, eigen2_arr, color='tab:green', linestyle='dotted', zorder=1, label='$\omega_2^2$')
plt.axhline(0, linewidth=1, color='black')
plt.axvline(3224, linestyle='dashed', linewidth=1, color='black')
plt.xlabel('Central Pressure [$MeV^4$]')
plt.ylabel('arcsinh($\omega_n^2 / \Omega^2$)')
plt.text(2.7,-1.3 ,'a', fontsize='13')
plt.text(1500,-1.3 ,'b', fontsize='13')
plt.text(3.6e6,-1.3 ,'c', fontsize='13')
plt.text(9e6,-1.3 ,'d', fontsize='13')
plt.xlim(1e-3)
plt.xscale('log')
plt.title('Eigenvalue Plot')
plt.ylim(-18)
plt.legend(frameon=True, edgecolor='black', fontsize=10)
plt.show()
```

```python
presscd_arr = np.genfromtxt('presscd_data.dat', dtype=float, skip_header=1)
freqcd_arr = np.genfromtxt('freqcd_data.dat', dtype=float, skip_header=1)
stabcd_arr = np.genfromtxt('stabcd_data.dat', dtype=float, skip_header=1)

angcd_arr = np.arcsinh((2*np.pi*freqcd_arr)**2)
eigencd_arr = angcd_arr*stabcd_arr

presscd1_arr = np.genfromtxt('presscd1_data.dat', dtype=float, skip_header=1)
freqcd1_arr = np.genfromtxt('freqcd1_data.dat', dtype=float, skip_header=1)
freqcd2_arr = np.genfromtxt('freqcd2_data.dat', dtype=float, skip_header=1)

angcd1_arr = np.arcsinh((2*np.pi*freqcd1_arr)**2)
eigencd1_arr = angcd1_arr

angcd2_arr = np.arcsinh((2*np.pi*freqcd2_arr)**2)
eigencd2_arr = angcd2_arr
```

```python
plt.plot(presscd_arr, eigencd_arr, zorder=3, label='$\omega_0^2$')
plt.plot(presscd1_arr, eigencd1_arr, color='tab:orange', linestyle='dashed',zorder=2, label='$\omega_1^2$')
plt.plot(presscd1_arr, eigencd2_arr, color='tab:green', linestyle='dotted',zorder=1, label='$\omega_2^2$')
plt.xscale('log')
plt.ylim(-14)
plt.xlim(3e6,2e7)
plt.text(6.4e6,-1.4 ,'c', fontsize='13')
plt.text(8.2e6,-1.4,'d', fontsize='13')
plt.title('Eigenvalue Plot')
plt.xlabel('Central Pressure [$MeV^4$]')
plt.ylabel('arcsinh($\omega_n^2 / \Omega^2$)')
plt.legend(frameon=True, edgecolor='black', fontsize=10)
plt.axhline(0, color='black')
plt.show()
```
